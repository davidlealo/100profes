### Matplotlib Visualization with Objects

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig0=plt.figure()
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0=plt.figure()
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='purple', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
axes0.set_xlim([0, 5])
axes0.set_ylim([-2, 2])
plt.show()
```

```python
fig0=plt.figure(figsize=(10, 1))
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0=plt.figure(figsize=(5, 2), dpi=100)
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='orange', linestyle='--', label='Sin')
axes0.plot(x, z, color='purple', linestyle='-.', label='Cos')
axes0.legend(loc=0)
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0=plt.figure(figsize=(8, 3), dpi=80)
axes1 = fig0.add_axes([0, 0, 0.3, 1])
axes2 = fig0.add_axes([0.5, 0, 0.3, 1])
axes1.plot(x, y, color='red', linestyle=':')
axes2.plot(x, z, color='blue', linestyle='-.')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')
plt.show()
```

```python
fig0=plt.figure(figsize=(8, 3), dpi=80)
axes1 = fig0.add_axes([0, 0, 0.2, 1])
axes2 = fig0.add_axes([0.5, 0, 0.2, 1])
axes1.plot(x, z, linestyle='none', marker='^', markersize=10, markerfacecolor='purple', markeredgecolor='red', alpha=0.5)
axes2.plot(x, y, color='blue', linestyle='--')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('COSINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('SInE')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# (0,0)
c = np.random.randn(1000)
axes[0,0].hist(c, bins=100, color='green', density=True, alpha=0.3)
axes[0,0].set_title('HISTOGRAM')
#(1,0)
axes[1,0].plot(x, z, linestyle='none', marker='^', markersize=10, markerfacecolor='purple', markeredgecolor='red', alpha=0.5)
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('COSINE')
#(0,1)
a = np.array(['Q1', 'Q2', 'Q3', 'Q4'])
b = np.array([234.0, 254.7, 144.6, 317.6])
axes[0,1].bar(a, b, color='blue', alpha=0.5)
axes[0,1].set_title('BAR PLOT')
#(1,1)
axes[1,1].plot(x, y, color='green', linestyle=':')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('SINE')

plt.tight_layout()
plt.show()
```

### Seaborn Visualization Library (with more examples)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
dat = sns.load_dataset('iris')
```

```python
sns.pairplot(dat)
plt.show()
```

```python
sns.pairplot(dat, hue='species')
plt.show()
```

```python
sns.pairplot(dat, hue='species', palette='coolwarm')
plt.show()
```

```python
g = sns.PairGrid(dat)
g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(dat)
g.map_diag(sns.distplot, color='orange')
g.map_upper(plt.scatter, color='green')
g.map_lower(sns.kdeplot, color='blue')
plt.show()
```

```python
g = sns.FacetGrid(dat, col='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
g = sns.FacetGrid(dat, row='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
g = sns.FacetGrid(dat, col='species')
g.map(plt.scatter, 'petal_length', 'petal_width', color='orange')
plt.show()
```
