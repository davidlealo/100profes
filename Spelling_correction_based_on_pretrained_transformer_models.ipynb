```python
from transformers import pipeline

fix_spelling = pipeline("text2text-generation",model="oliverguhr/spelling-correction-english-base")

print(fix_spelling("lets do a comparsion",max_length=2048))
```

```python
print(fix_spelling("get does a comparision",max_length=2048))
```

```python
print(fix_spelling("get does am comparision",max_length=2048))
```

```python
print(fix_spelling("geting does an comparsion",max_length=2048))
```
