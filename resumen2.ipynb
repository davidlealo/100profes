```python
import os
import anthropic
from anthropic import Anthropic

# [Las funciones request_anthropic_api_key() y store_anthropic_api_key() se mantienen igual]

def chat_with_anthropic():
    if "ANTHROPIC_API_KEY" not in os.environ:
        print("API key no encontrada. Por favor, vuelva a ingresar su clave.")
        return

    # Inicializar el cliente con la API key
    client = Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

    print("\nInteractúa con el chatbot. Escribe 'salir' para finalizar.\n")

    while True:
        user_input = input("Tú: ")

        if user_input.lower() == "salir":
            print("¡Gracias por usar el chatbot! Hasta luego.")
            break

        try:
            # Crear el mensaje usando el nuevo formato
            message = client.messages.create(
                model="claude-3-5-sonnet-20241022",
                max_tokens=100,
                temperature=0.7,
                system="Eres un modelo que ayuda al usuario a mejorar su productividad. Realiza preguntas y colabora con las respuestas que te pida. En especial ayúdale a hacer resumenes de textos",
                messages=[
                    {"role": "user", "content": user_input}
                ]
            )

            # Obtener la respuesta
            ai_response = message.content[0].text
            print(f"AI: {ai_response.strip()}")

        except Exception as e:
            print(f"Error: {e}")
            break
```

## Programa de resumen

Primera versión que también exporta el texto original

```python
import re
from typing import Optional
from IPython.display import display, HTML
import ipywidgets as widgets
from google.colab import files
import io
import PyPDF2
from anthropic import Anthropic
import os
from docx import Document
import datetime

class PDFDocumentSummarizer:
    def __init__(self, words_per_page: int = 250):
        self.words_per_page = words_per_page
        self.client = None
        if "ANTHROPIC_API_KEY" in os.environ:
            self.client = Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

    def extract_text_from_pdf(self, pdf_content: bytes) -> str:
        """Extrae texto de un archivo PDF"""
        try:
            pdf_file = io.BytesIO(pdf_content)
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
            return text
        except Exception as e:
            raise Exception(f"Error al procesar PDF: {str(e)}")

    def count_words(self, text: str) -> int:
        """Cuenta palabras en el texto"""
        return len(text.split())

    def estimate_pages(self, text: str) -> float:
        """Estima número de páginas basado en el conteo de palabras"""
        return self.count_words(text) / self.words_per_page

    def summarize(self, text: str, max_length_ratio: float = 0.3) -> str:
        """Crea un resumen del texto manteniendo la información clave"""
        # Limpieza básica del texto
        text = re.sub(r'\s+', ' ', text).strip()

        # Dividir en oraciones (mejorado para manejar múltiples puntuaciones)
        sentences = [s.strip() for s in re.split(r'(?<=[.!?])\s+', text) if s.strip()]

        target_sentences = int(len(sentences) * max_length_ratio)

        if target_sentences >= len(sentences):
            return text

        summary_sentences = []

        # Incluir primera oración
        summary_sentences.append(sentences[0])

        # Seleccionar oraciones distribuidas uniformemente
        if target_sentences > 2:
            step = (len(sentences) - 2) // (target_sentences - 2)
            for i in range(1, target_sentences - 1):
                idx = i * step
                if idx < len(sentences):
                    summary_sentences.append(sentences[idx])

        # Incluir última oración
        if target_sentences > 1:
            summary_sentences.append(sentences[-1])

        return ' '.join(summary_sentences)

    def process_document(self, text: str, max_pages: int) -> str:
        """Procesa el documento y crea un resumen dentro de los límites de páginas"""
        current_pages = self.estimate_pages(text)

        if current_pages <= max_pages:
            return text

        target_ratio = max_pages / current_pages
        summary = self.summarize(text, target_ratio)

        return summary

    def enhance_summary_with_anthropic(self, text: str) -> str:
        """Mejora el resumen usando Claude"""
        if not self.client:
            raise Exception("API key de Anthropic no configurada")

        try:
            message = self.client.messages.create(
                model="claude-3-5-sonnet-20241022",
                max_tokens=1500,
                temperature=0.7,
                system="Eres un experto en resumir y mejorar textos. Tu tarea es mejorar el resumen proporcionado, "
                       "manteniendo la información clave pero mejorando la claridad, coherencia y fluidez.",
                messages=[
                    {"role": "user",
                     "content": f"Por favor, mejora el siguiente resumen manteniendo la información esencial pero "
                               f"haciéndolo más claro y coherente:\n\n{text}"}
                ]
            )
            return message.content[0].text
        except Exception as e:
            raise Exception(f"Error al procesar con Anthropic: {str(e)}")

    def save_to_docx(self, original_text: str, summary: str, enhanced_summary: str, filename: str) -> str:
        """Guarda los resúmenes en un archivo DOCX"""
        doc = Document()

        # Título
        doc.add_heading('Informe de Resumen', 0)

        # Fecha y hora
        doc.add_paragraph(f'Generado: {datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')

        # Texto original
        doc.add_heading('Texto Original', level=1)
        doc.add_paragraph(original_text)

        # Resumen inicial
        doc.add_heading('Resumen Inicial', level=1)
        doc.add_paragraph(summary)

        # Resumen mejorado
        doc.add_heading('Resumen Mejorado por Claude', level=1)
        doc.add_paragraph(enhanced_summary)

        # Estadísticas
        doc.add_heading('Estadísticas', level=1)
        stats = doc.add_paragraph()
        stats.add_run(f'Palabras en texto original: {self.count_words(original_text)}\n')
        stats.add_run(f'Palabras en resumen inicial: {self.count_words(summary)}\n')
        stats.add_run(f'Palabras en resumen mejorado: {self.count_words(enhanced_summary)}\n')

        # Guardar archivo
        output_filename = f'resumen_{filename.replace(".pdf", "")}.docx'
        doc.save(output_filename)
        return output_filename

def create_interface():
    # Crear widgets
    status_label = widgets.HTML(value="<h3>Estado: Esperando archivo...</h3>")
    text_area = widgets.Textarea(
        value='',
        placeholder='El texto extraído del PDF aparecerá aquí...',
        description='Texto:',
        layout={'width': '100%', 'height': '200px'},
        disabled=True
    )
    summary_area = widgets.Textarea(
        value='',
        placeholder='El resumen aparecerá aquí...',
        description='Resumen:',
        layout={'width': '100%', 'height': '200px'},
        disabled=True
    )
    enhanced_summary_area = widgets.Textarea(
        value='',
        placeholder='El resumen mejorado por Claude aparecerá aquí...',
        description='Resumen mejorado:',
        layout={'width': '100%', 'height': '200px'},
        disabled=True
    )
    max_pages = widgets.IntText(
        value=1,
        description='Máx páginas:',
        min=1
    )
    output = widgets.Output()

    def on_upload_click(b):
        with output:
            output.clear_output()
            status_label.value = "<h3>Estado: Subiendo archivo...</h3>"
            try:
                uploaded = files.upload()
                for filename, content in uploaded.items():
                    if not filename.lower().endswith('.pdf'):
                        status_label.value = "<h3>Estado: Error - El archivo debe ser PDF</h3>"
                        return

                    summarizer = PDFDocumentSummarizer()

                    # Extraer texto
                    status_label.value = "<h3>Estado: Procesando PDF...</h3>"
                    extracted_text = summarizer.extract_text_from_pdf(content)
                    text_area.value = extracted_text

                    # Generar resumen inicial
                    status_label.value = "<h3>Estado: Generando resumen inicial...</h3>"
                    initial_summary = summarizer.process_document(extracted_text, max_pages.value)
                    summary_area.value = initial_summary

                    # Mejorar con Claude
                    status_label.value = "<h3>Estado: Mejorando resumen con Claude...</h3>"
                    enhanced_summary = summarizer.enhance_summary_with_anthropic(initial_summary)
                    enhanced_summary_area.value = enhanced_summary

                    # Guardar documento
                    status_label.value = "<h3>Estado: Guardando documento...</h3>"
                    output_filename = summarizer.save_to_docx(
                        extracted_text, initial_summary, enhanced_summary, filename
                    )

                    status_label.value = f"<h3>Estado: ¡Proceso completado! Archivo guardado como: {output_filename}</h3>"

                    print("Estadísticas:")
                    print(f"- Palabras en texto original: {summarizer.count_words(extracted_text)}")
                    print(f"- Palabras en resumen inicial: {summarizer.count_words(initial_summary)}")
                    print(f"- Palabras en resumen mejorado: {summarizer.count_words(enhanced_summary)}")
                    print(f"- Archivo guardado como: {output_filename}")

            except Exception as e:
                status_label.value = f"<h3>Estado: Error - {str(e)}</h3>"

    upload_btn = widgets.Button(description='Cargar PDF')
    upload_btn.on_click(on_upload_click)

    # Layout
    controls = widgets.HBox([upload_btn, max_pages])

    display(HTML("<h2>Resumidor de PDFs con mejora por Claude</h2>"))
    display(status_label)
    display(controls)
    display(widgets.HTML(value="<h4>Texto extraído del PDF:</h4>"))
    display(text_area)
    display(widgets.HTML(value="<h4>Resumen inicial:</h4>"))
    display(summary_area)
    display(widgets.HTML(value="<h4>Resumen mejorado por Claude:</h4>"))
    display(enhanced_summary_area)
    display(output)

# Ejecutar la interfaz
create_interface()
```
